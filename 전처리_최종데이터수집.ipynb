# 클라이밍(climbing)

```python
###각자정리한 데이터를 바탕으로 위도경도 따기###

#사용한 라이브러리 임포트
import json           
import requests         #카카오 API를 위한 라이브러리

import numpy as np
import pandas as pd 
import csv                                #CSV 파일에서 읽어 들인 데이터는 모두 문자로 인식됩니다. ★
```

```python
#구글 드라이브와 연동(파일 경로 체크)

from google.colab import drive
drive.mount('/content/gdrive')
directory_path = '/content/gdrive/My Drive/22-1파이썬_팀프로젝트/place_data/before_csv/'

#*********<파일이름만 변경하기>************
f = open(directory_path + 'climbing.csv', 'r', encoding='utf8') 
data = csv.reader(f)                      # 파일 읽기
next(data)                                # 표의 헤더 제거하기
data = list(data)                         # 리스트로 변환하기
```

```python
# 넘파이 배열로 변환
data = np.array(data)
data = np.delete(data,2,axis=1)
data = np.delete(data,2,axis=1)
print(data)
```

```python
print(data.shape)
print(data.ndim)
```

```python
#카카오 API 사용하여 주소로부터 위도경도 정보 얻어내기

def addr_to_lat_lon(addr):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={address}'.format(address=addr)
    headers = {"Authorization": "KakaoAK " + 'fd32761e06a5fcb3126301f3e1e9d69a'}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address']
    return float(match_first['x']), float(match_first['y'])
```

```python
# 카카오API 함수와 반복문을 사용해 '위도'와'경도' 열을 얻기
global location
location = np.empty((len(data),2),float)

for i in range(len(data)):
  try :
    location[i] = np.array(addr_to_lat_lon(data[i,1]))
  except:
    location[i]=np.NaN
  print(i,data[i,1],location[i],data[i,1])
    
```

```python
pd.options.display.float_format = '{: .8f}'.format  #위도경도데이터 소수점8자리로 출력
```

```python
# 위도,경도 데이터프레임 만들기

df1 = pd.DataFrame(location,columns=['경도','위도'])

df1.head(5)      #5개만 미리보기
```

```python
# 장소명,주소 데이터 프레임 만들기

df2 = pd.DataFrame(data, columns = ['장소명','주소'])
df2.head(5)
```

```python
# 위도경도 데이터프레임과 장소명 주소 데이터프레임 합치기
climbing = df2.join(df1)   
# 분류 컬럼 추가
climbing['분류'] = 'climbing'

climbing.head(5)
```

```python
climbing.isnull().sum() #결측치(NAN)값 개수 확인
```

```python
climbing.dropna(inplace=True)     #결측치(NaN값) 제거해서 저장
```

```python
climbing.head(5)
```

```python
climbing.to_csv('/content/gdrive/My Drive/22-1파이썬_팀프로젝트/place_data/csv/climbing.csv', index = None)
```
